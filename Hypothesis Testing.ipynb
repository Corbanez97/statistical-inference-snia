```python
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(6, 10), dpi=100, facecolor='w', edgecolor='k')
```

```python
def normal(mu, sigma, x):
    return np.exp(-0.5*((x - mu)/sigma)**2)/(2*sigma*np.sqrt(2*np.pi))
```

```python
x = np.linspace(-10, 10, 1000)
xright = np.linspace(0.1, 10, 1000)
xleft = np.linspace(-10, 0.1, 1000)
```

```python
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(4)

plt.axvline(
    x = 0.1, 
    color = 'k', 
    label = '$x_c$', 
    alpha = 0.5
)

plt.plot(
    x, 
    normal(-1.0, 0.5, x), 
    '-.', 
    label = '$H_0$'
)
plt.plot(
    x, 
    normal(0.9, 0.6, x), 
    '--', 
    label = '$H_A$'
)

plt.fill_between(
    xright, 
    normal(-1.0, 0.5, xright), 
    alpha = 0.3, 
    label = 'Type I Error'
)
plt.fill_between(
    xleft, 
    normal(0.9, 0.6, xleft), 
    alpha = 0.3, 
    label = 'Typer II Error'
)

plt.axis('off')
plt.xlim([-3, 3])
plt.legend(fontsize=14)

plt.savefig('images/alpha_beta_errors.pdf')  
```
