```python
import matplotlib.pyplot as plt
from pylab import rcParams

import numpy as np
import scipy.stats as stats

import pickle
```

```python
with open('data/multiprocessing_log.pickle', 'rb') as handle:
    log = pickle.load(handle)
```

```python
rcParams['figure.figsize'] = 10, 10

sigma1, sigma2, sigma3 = 0.6827, 0.9545, 0.9973
lambda1, lambda2, lambda3 = stats.chi2.ppf(sigma1, 2), stats.chi2.ppf(sigma2, 2), stats.chi2.ppf(sigma3, 2)
for grid_size in log:
    for item in log[grid_size]['result']:
        if item[1] < lambda1:
            plt.scatter(item[0][0], item[0][1], s = 1, marker = 'x', color = 'red', label = '$3\sigma$')
        elif lambda1 < item[1] < lambda2:
            plt.scatter(item[0][0], item[0][1], s = 1, marker = 'x', color = 'blue', label = '$3\sigma$')
        elif lambda2 < item[1] < lambda3:
            plt.scatter(item[0][0], item[0][1], s = 1, marker = 'x', color = 'green', label = '$3\sigma$')
        elif lambda3 < item[1]:
            plt.scatter(item[0][0], item[0][1], s = 0.14, alpha = 0.24, marker = 'x', color = 'grey', label = '$3\sigma$')

    plt.xlabel('$\Omega_{\Lambda}$')
    plt.ylabel('$\Omega_m$')
    plt.title(grid_size)
    plt.show()
```
